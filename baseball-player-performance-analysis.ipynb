* The chart shows the percentage of missing data for features in the dataset, with ExitVelocity and LaunchAngle having the highest missing percentages at around 66.88% and 66.89%, respectively. The other features (Velocity, HBreak, VBreak, PlateX, and PlateZ) have minimal missing values, all below 1%. 
* For the features ExitVelocity and LaunchAngle, missing values are expected and reasonable because it will only have values when the batter doesn't make contact with the ball. Otherwise, if Batter doesn't swing, it wouldn't have records. So we don't need to do any replace for the missing data. If these features have values, it indicates that the batter made contact, either resulting in a foul ball or a ball put into play. In our dataset, we only have the feature IsInPlay to indicate whether the ball was put into play, but we lack a specific indicator for foul balls. However, we can create a new feature, named isHit, based on whether ExitVelocity and LaunchAngle have values. If both ExitVelocity and LaunchAngle are non-null, we can set isHit to 1, representing that the batter made contact.

```python

features_with_missing = ['Velocity', 'HBreak', 'VBreak', 'PlateX', 'PlateZ']

plt.figure(figsize=(12, 10))

for i, feature in enumerate(features_with_missing):
    plt.subplot(4, 2, i + 1)
    plt.hist(data_new[feature].dropna(), bins=50, color='blue', alpha=0.7)  
    plt.title(f'Distribution of {feature} (Skewness: {skew(data_new[feature].dropna()):.2f})')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

* If skewness is close to 0 (normal distribution), the mean is typically a good choice for replacing missing values.
If skewness is far from 0 (highly skewed), the median is preferred because it is less affected by outliers and skewed distributions.
* Therefore, the feature Velocity and VBreak are highly skewed, the missing value can be replaced by using Median. Other features are replaced by using Mean. 

```python
# Replace missing values based on skewness
data_new['Velocity'].fillna(data_new['Velocity'].median(), inplace=True)  # Skewed, use median
data_new['HBreak'].fillna(data_new['HBreak'].mean(), inplace=True)  # Fairly symmetric, use mean
data_new['VBreak'].fillna(data_new['VBreak'].median(), inplace=True)  # Skewed, use median
data_new['PlateX'].fillna(data_new['PlateX'].mean(), inplace=True)  # Symmetric, use mean
data_new['PlateZ'].fillna(data_new['PlateZ'].mean(), inplace=True)  # Symmetric, use mean
```

* For the purpose of Question 1 and Question 2, where we are focused on evaluating the batter's contact skill, I will calculate contact skill using the ratio of IsInPlay to IsSwing. This approach disregards foul balls, focusing only on the batter's ability to put the ball into play. However, I will retain the isHit feature for potential future analyses involving other aspects of batting performance.

## Define Contact Skill

* Define the Contact Skill based on the IsSwing and IsInPlay, the Contact rate can be caculated by:  Contact Skill (ContactRate) =  IsInPlay / IsSwing  

```python
# Calculate contact skill for Batter 1 in 2022 and 2023
total_swings_batter_1_2022 = batter_1_data[(batter_1_data['Year'] == 2022)]['IsSwing'].sum()
total_in_play_batter_1_2022 = batter_1_data[(batter_1_data['Year'] == 2022)]['IsInPlay'].sum()
batter_1_contact_skill_2022 = total_in_play_batter_1_2022 / total_swings_batter_1_2022

total_swings_batter_1_2023 = batter_1_data[(batter_1_data['Year'] == 2023)]['IsSwing'].sum()
total_in_play_batter_1_2023 = batter_1_data[(batter_1_data['Year'] == 2023)]['IsInPlay'].sum()
batter_1_contact_skill_2023 = total_in_play_batter_1_2023 / total_swings_batter_1_2023

# Calculate contact skill for Batter 2 in 2022 and 2023
total_swings_batter_2_2022 = batter_2_data[(batter_2_data['Year'] == 2022)]['IsSwing'].sum()
total_in_play_batter_2_2022 = batter_2_data[(batter_2_data['Year'] == 2022)]['IsInPlay'].sum()
batter_2_contact_skill_2022 = total_in_play_batter_2_2022 / total_swings_batter_2_2022

total_swings_batter_2_2023 = batter_2_data[(batter_2_data['Year'] == 2023)]['IsSwing'].sum()
total_in_play_batter_2_2023 = batter_2_data[(batter_2_data['Year'] == 2023)]['IsInPlay'].sum()
batter_2_contact_skill_2023 = total_in_play_batter_2_2023 / total_swings_batter_2_2023
```

```python

# Define the contact skills data for each batter
batters = ['Batter 1', 'Batter 2']
contact_skills_2022 = [batter_1_contact_skill_2022, batter_2_contact_skill_2022]
contact_skills_2023 = [batter_1_contact_skill_2023, batter_2_contact_skill_2023]
plt.figure(figsize=(6, 3))
bar_width = 0.35
x = np.arange(len(batters))

# Create the bar chart with two bars for each batter (one for 2022, one for 2023)
plt.bar(x - bar_width/2, contact_skills_2022, width=bar_width, label='2022', color='Black')
plt.bar(x + bar_width/2, contact_skills_2023, width=bar_width, label='2023', color='Red')

plt.xlabel('Batter', fontsize=14)
plt.ylabel('Contact Skill', fontsize=14)
plt.title('Comparison of Contact Skill for Batter 1 and Batter 2 (2022 vs 2023)', fontsize=16, fontweight='bold')
plt.xticks(x, batters, fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

* The bar chart shows the comparison of contact skill for Batter 1 and Batter 2 between 2022 and 2023. For Batter 1, the contact skill improved significantly (+10%) in 2023, with a higher rate compared to 2022. In contrast, Batter 2 experienced a decline (-5%) in contact skill in 2023 compared to their performance in 2022. 

```python

# Combine Batter 1 and Batter 2 data for 2022 and 2023 for further analysis
grouped_data_batter_1 = batter_1_data[(batter_1_data['Year'] == 2022) | (batter_1_data['Year'] == 2023)]
grouped_data_batter_2 = batter_2_data[(batter_2_data['Year'] == 2022) | (batter_2_data['Year'] == 2023)]

# Function to calculate contact rate by binning for numerical features
def calculate_contact_rate_per_bin(grouped_data, feature):
    grouped_data_copy = grouped_data.copy()
    grouped_data_copy[feature + '_bins'] = pd.cut(grouped_data_copy[feature], bins=8)
    contact_rate_data = grouped_data_copy.groupby([feature + '_bins', 'Year']).agg(
        total_swing=('IsSwing', 'sum'),
        total_inplay=('IsInPlay', 'sum')
    )
    contact_rate_data['contact_rate'] = contact_rate_data['total_inplay'] / contact_rate_data['total_swing']
    contact_rate_data = contact_rate_data[contact_rate_data['total_swing'] > 0]
    return contact_rate_data['contact_rate'].unstack()

# Function to plot contact rate by feature bins
def plot_filtered_contact_rate_per_bin(grouped_data, feature, batter_label, ax):
    
    contact_rate_per_bin = calculate_contact_rate_per_bin(grouped_data, feature)
    contact_rate_filtered = contact_rate_per_bin.dropna()
    contact_rate_filtered.plot(kind='bar', ax=ax, title=f'Contact Rate by {feature} Bins for {batter_label}', 
                               figsize=(8, 6), color=['black', 'red'])  
    ax.set_ylabel('Contact Rate (IsInPlay)')
    ax.set_xlabel(f'{feature} Bins')
    ax.set_xticklabels([f'({bin.left:.2f}, {bin.right:.2f}]' for bin in contact_rate_filtered.index], rotation=45)
    ax.legend(title="Year")

# Function to plot contact rate for the categorical feature (PitcherHand)
def plot_contact_rate_by_pitcher_hand(grouped_data, batter_label, ax):
    contact_rate_pitcher_hand = grouped_data.groupby(['PitcherHand', 'Year']).agg(
        total_swing=('IsSwing', 'sum'),
        total_inplay=('IsInPlay', 'sum')
    )
    contact_rate_pitcher_hand['contact_rate'] = contact_rate_pitcher_hand['total_inplay'] / contact_rate_pitcher_hand['total_swing']
    contact_rate_pitcher_hand['contact_rate'].unstack().plot(kind='bar', ax=ax, title=f'Contact Rate by PitcherHand for {batter_label}', 
                                                             figsize=(8, 6), color=['black', 'red']) 
    ax.set_ylabel('Contact Rate (IsInPlay)')
    ax.set_xlabel('Pitcher Hand (L: Left, R: Right)')
    ax.legend(title="Year")

# Create a function to handle both numerical and categorical features for both batters
def plot_contact_rate_for_batters(feature):
    fig, axs = plt.subplots(2, 1, figsize=(10, 16))

    if feature == 'PitcherHand':
        # Plot for categorical feature (PitcherHand)
        plot_contact_rate_by_pitcher_hand(grouped_data_batter_1, 'Batter 1', axs[0])
        plot_contact_rate_by_pitcher_hand(grouped_data_batter_2, 'Batter 2', axs[1])
    else:
        # Plot for numerical feature with bins
        plot_filtered_contact_rate_per_bin(grouped_data_batter_1, feature, 'Batter 1', axs[0])
        plot_filtered_contact_rate_per_bin(grouped_data_batter_2, feature, 'Batter 2', axs[1])

    plt.tight_layout()
    plt.show()

# Features to choose from
features = ['Velocity', 'HBreak', 'VBreak', 'PlateX', 'PlateZ','PitcherHand']

# Create a dropdown widget
dropdown = widgets.Dropdown(
    options=features,
    value='Velocity',
    description='Feature:',
    disabled=False,
)
interact(plot_contact_rate_for_batters, feature=dropdown)
```

* In the 2023 season, Batter 1 demonstrated a significant increase in contact rate for pitches with velocities between 77 and 102 mph, suggesting improved ability to handle faster pitches, which are often fastballs. This indicates that Batter 1 has adapted well to facing high-speed pitches, making them a more formidable opponent against pitchers who rely heavily on fastballs. Additionally, the increased contact rate for horizontal and vertical breaks (HBreak and VBreak) implies that Batter 1 has also improved significantly in handling breaking balls, showcasing overall growth in pitch adaptability. In contrast, Batter 2's decreased contact rate, particularly for moderate to high velocity pitches, suggests difficulty in keeping up with faster pitches, making them potentially more vulnerable to power pitchers. Furthermore, Batter 1's improved performance against both left- and right-handed pitchers highlights better adaptability across game situations, whereas Batter 2's decline against left-handed pitchers indicates a persistent weakness that opposing teams could continue to target.

# Question2 - Statistical Significance

## Statistical Comparison

### Z-test 

```python

# Perform Z-test for Batter 1
count_batter_1 = [total_in_play_batter_1_2022, total_in_play_batter_1_2023]  # Successful contacts (InPlay)
nobs_batter_1 = [total_swings_batter_1_2022, total_swings_batter_1_2023]  # Total swings (IsSwing)

z_stat_batter_1, p_value_batter_1 = proportions_ztest(count_batter_1, nobs_batter_1)

# Perform Z-test for Batter 2
count_batter_2 = [total_in_play_batter_2_2022, total_in_play_batter_2_2023]  # Successful contacts (InPlay)
nobs_batter_2 = [total_swings_batter_2_2022, total_swings_batter_2_2023]  # Total swings (IsSwing)

z_stat_batter_2, p_value_batter_2 = proportions_ztest(count_batter_2, nobs_batter_2)


results_table = pd.DataFrame({
    'Batter': ['Batter 1', 'Batter 2'],
    'Z-statistic': [z_stat_batter_1, z_stat_batter_2],
    'P-value': [p_value_batter_1, p_value_batter_2]
})

(results_table.round(4)) 
```

* The Z-test result for Batter 1 shows a statistically significant difference in contact skill between 2022 and 2023, as indicated by the very low p-value (p < 0.0001), suggesting a notable change. For Batter 2, the p-value (0.5173) indicates no statistically significant difference in contact skill between the two years, suggesting that any observed variation could be due to random chance.

### Bayesian Analysis (posterior distribution)

```python
# Function to perform Bayesian analysis with a Beta prior
def bayesian_update(k, n, alpha_prior=1, beta_prior=1):
    alpha_post = alpha_prior + k
    beta_post = beta_prior + (n - k)
    return alpha_post, beta_post


def plot_posterior(alpha_post, beta_post, batter_name, year, color, linestyle='-'):
    x = np.linspace(0, 1, 100)
    y = beta.pdf(x, alpha_post, beta_post)
    plt.plot(x, y, label=f'{batter_name} {year}', color=color, linestyle=linestyle)
    plt.fill_between(x, y, color=color, alpha=0.2)

k_1_2022 = total_in_play_batter_1_2022
n_1_2022 = total_swings_batter_1_2022
k_1_2023 = total_in_play_batter_1_2023
n_1_2023 = total_swings_batter_1_2023
alpha_post_1_2022, beta_post_1_2022 = bayesian_update(k_1_2022, n_1_2022)
alpha_post_1_2023, beta_post_1_2023 = bayesian_update(k_1_2023, n_1_2023)

k_2_2022 = total_in_play_batter_2_2022
n_2_2022 = total_swings_batter_2_2022
k_2_2023 = total_in_play_batter_2_2023
n_2_2023 = total_swings_batter_2_2023
alpha_post_2_2022, beta_post_2_2022 = bayesian_update(k_2_2022, n_2_2022)
alpha_post_2_2023, beta_post_2_2023 = bayesian_update(k_2_2023, n_2_2023)


plt.figure(figsize=(7, 3))
plot_posterior(alpha_post_1_2022, beta_post_1_2022, 'Batter 1', '2022', color="black")
plot_posterior(alpha_post_1_2023, beta_post_1_2023, 'Batter 1', '2023', color="black", linestyle='--')
plot_posterior(alpha_post_2_2022, beta_post_2_2022, 'Batter 2', '2022', color="red")
plot_posterior(alpha_post_2_2023, beta_post_2_2023, 'Batter 2', '2023', color="red", linestyle='--')
plt.xlabel('Contact Rate')
plt.ylabel('Density')
plt.title('Posterior Distributions of Contact Rates for Both Batters (2022 vs 2023)')

plt.text(0.2, 28, 'Batter 1 is statistically significant', color='black', fontsize=10)
plt.text(0.05,12, 'Batter 2 (red) is not statistically significant', color='red', fontsize=10)

plt.legend()
plt.show()
```

* The posterior distributions indicate that Batter 1's contact rate in 2023 (dashed black line) increased compared to 2022 (solid black line), as evidenced by the shift of the peak to the right. Conversely, Batter 2's contact rate remained fairly consistent between 2022 (solid red line) and 2023 (dashed red line), with similar peak positions and overlapping distributions, indicating minimal change in contact skill for Batter 2.

## Evaluate PitchTypeCode

* The PitchTypeCode is important because it helps differentiate the types of pitches thrown, such as fastballs, sliders, or curveballs, which can have a significant impact on a batter's performance and contact skill. Understanding how a batter performs against different pitch types provides valuable insights for tailoring pitching strategies and improving batting techniques.
* To expedite the notebook's execution and demonstrate the process, I used only 10,000 data points from the historical data to train and test the Predictive multiple classification PitchTypeCode model. For better model performance and predictive power, you can use the entire dataset instead.

```python
pitch_type_distribution = historical_data['PitchTypeCode'].value_counts(normalize=True) * 100

encoder = OneHotEncoder(sparse=False)
pitcher_hand_encoded = encoder.fit_transform(historical_data[['PitcherHand']])
pitcher_hand_encoded_df = pd.DataFrame(pitcher_hand_encoded, columns=encoder.get_feature_names_out(['PitcherHand']))
historical_data = pd.concat([historical_data, pitcher_hand_encoded_df], axis=1)
historical_data.drop(columns=['PitcherHand'], inplace=True)

sampled_data = historical_data.sample(n=10000, random_state=42)
features = ['PitcherHand_R', 'PitcherHand_L', 'PitchVelo', 'PitchHorizBreak', 'PitchVertBreak']
X = sampled_data[features]
y = sampled_data['PitchTypeCode']
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
X_imputed_df = pd.DataFrame(X_imputed, columns=features)


X_train, X_test, y_train, y_test = train_test_split(X_imputed_df, y, test_size=0.3, random_state=42)

classifier = RandomForestClassifier(random_state=42)
classifier.fit(X_train, y_train)
y_score = classifier.predict_proba(X_test)

lb = LabelBinarizer()
y_test_bin = lb.fit_transform(y_test)

n_classes = y_test_bin.shape[1]
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure(figsize=(10, 8))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {lb.classes_[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-AUC Curve for All Pitch Types')
plt.legend(loc='lower right')
plt.show()
```

* The ROC-AUC curve for all classes shows the performance of the predictive model for different pitch types. Types like CH (Changeup), CU (Curveball), FF (4-seam Fast), SL (Slider), and SI (Sinker) have high AUC scores ( 0.92), indicating that the model performs well in distinguishing these pitch types with high true positive rates and low false positive rates. However, class FA (2-seam Fast) has a relatively low AUC of 0.64, suggesting that the model struggles to accurately predict this pitch type. The overall results demonstrate strong model performance for most classes, but there is room for improvement in predicting certain pitch types with lower AUC scores like FA and UN (unknown). However, we can improve the performance by using all the historical dataset.

### Analyze Contact Rate for PitchType

```python
# Step 1: Map PitchTypeCode to Full Names
pitch_type_mapping = {
    'FF': 'Four-Seam Fastball (FF)',
    'SL': 'Slider (SL)',
    'CH': 'Changeup (CH)',
    'CU': 'Curveball (CU)',
    'FT': 'Two-Seam Fastball (FT)',
    'SI': 'Sinker (SI)',
    'FC': 'Cutter (FC)',
    'SW': 'Slow Curve (SW)',
    'FS': 'Split-Finger Fastball (FS)',
    'UN': 'Unknown (UN)',
    'KC': 'Knuckle Curve (KC)',
    'FA': 'Fastball (FA)',
    'SV': 'Screwball (SV)',
    'KN': 'Knuckleball (KN)',
    'SC': 'Screwball Curve (SC)',
    'FO': 'Forkball (FO)',
}

# Add a new column with full pitch type names
data_new['PitchTypeName'] = data_new['PredictedPitchType'].map(pitch_type_mapping)

# Step 2: Create a function to plot contact rate for the selected batter
def plot_contact_rate_by_pitchtype_with_percentage_change(batter_id):
    # Filter the data for the selected batter and years 2022 and 2023
    grouped_data = data_new[(data_new['BatterID'] == batter_id) & (data_new['Year'].isin([2022, 2023]))]
    if len(grouped_data) <= 300:
        print(f"Not enough data available for Batter {batter_id}. Only {len(grouped_data)} data points found.")
        return
    years_available = grouped_data['Year'].unique()
    if 2022 not in years_available or 2023 not in years_available:
        print(f"Data for both 2022 and 2023 is required for Batter {batter_id}. Only data for years {years_available} found.")
        return
    grouped_data_copy = grouped_data.copy()
    contact_skill_per_pitchtype = (
        grouped_data_copy.groupby(['PitchTypeName', 'Year']).apply(lambda x: x['IsInPlay'].sum() / x['IsSwing'].sum())
    ).unstack()
    contact_skill_filtered = contact_skill_per_pitchtype[(contact_skill_per_pitchtype[2022] > 0) & (contact_skill_per_pitchtype[2023] > 0)]
    contact_skill_filtered['PercentageChange'] = 100 * (contact_skill_filtered[2023] - contact_skill_filtered[2022])

    ax = contact_skill_filtered[[2022, 2023]].plot(
        kind='bar',
        figsize=(10, 3),
        color=['black', 'red'], 
        title=f'Contact Skill by Pitch Type for Batter {batter_id}'
    )
    for i, (index, row) in enumerate(contact_skill_filtered.iterrows()):
        pct_change = row['PercentageChange']
        color = 'red' if pct_change > 0 else 'blue' 
        ax.text(i, max(row[2022], row[2023]), f'{pct_change:.2f}%', ha='center', va='bottom', fontsize=10, color=color)

    plt.ylim(0, 1)  
    plt.ylabel('Contact Skill (InPlay / Swing)')
    plt.xlabel('Pitch Type')
    plt.grid(True)
    plt.title(f'Contact Skill by Pitch Type for Batter {batter_id}', pad=20) 
    plt.show()

# Step 3: Create an interactive widget to choose between all available batters
batter_ids = data_new['BatterID'].unique()
interact(plot_contact_rate_by_pitchtype_with_percentage_change, batter_id=widgets.Dropdown(
    options=batter_ids,
    value=1,
    description='Select Batter:',
    style={'description_width': 'initial'}
));
```

* The chart shows the change in contact skill for Batter 1 across different pitch types between the 2022 and 2023 seasons. For most pitch types, Batter 1's contact skill improved, with significant increases for the Cutter (FC) by 21.83%, Split-Finger Fastball (FS) by 13.33%, and Changeup (CH) by 11.71%. This analysis is powerful as it allows us to switch between different batters and evaluate each batter's performance against various pitch types. This insight can help coaches and analysts determine individual strengths and weaknesses for each batter, enabling more effective training, game strategy, and adjustments in batting approaches to maximize success against specific pitchers.

## Evaluate the HotZone

* The goal of Hitting HotZone analysis is to visualize the contact rate of different batters in specific areas (based on the feature, PlateX and PlateZ) of the strike zone, helping identify their "hot" and "cold" zones for making successful contact. By allowing the selection of different batters, this tool provides a powerful way to tailor training strategies and make informed decisions about batting approaches, which can ultimately improve performance against various pitchers.

```python
# Define a function to create heatmaps based on selected batterID
def plot_heatmaps(batterID):
    # Filter data for the selected batter and only swing attempts
    batter_data_2022 = data_new[(data_new['BatterID'] == batterID) & (data_new['Year'] == 2022) & (data_new['IsSwing'] == 1)]
    batter_data_2023 = data_new[(data_new['BatterID'] == batterID) & (data_new['Year'] == 2023) & (data_new['IsSwing'] == 1)]

    # Check if there are any data points available for 2022 or 2023
    if batter_data_2022.empty or batter_data_2023.empty:
        print(f"Not enough data available for Batter {batterID} in either 2022 or 2023.")
        return

    # Define bins for PlateX and PlateZ (adjust the bin size)
    x_bins = np.linspace(-2, 2, 4)  # PlateX ranges from -2 to 2
    z_bins = np.linspace(1, 5, 4)   # PlateZ ranges from 1 to 5 (strike zone height)

    # Create pivot table for 2022
    pivot_table_2022 = batter_data_2022.pivot_table(index=pd.cut(batter_data_2022['PlateZ'], z_bins), 
                                                     columns=pd.cut(batter_data_2022['PlateX'], x_bins), 
                                                     values='IsMiss', 
                                                     aggfunc=lambda x: 1 - x.mean()).fillna(0)  

    # Create pivot table for 2023
    pivot_table_2023 = batter_data_2023.pivot_table(index=pd.cut(batter_data_2023['PlateZ'], z_bins), 
                                                     columns=pd.cut(batter_data_2023['PlateX'], x_bins), 
                                                     values='IsMiss', 
                                                     aggfunc=lambda x: 1 - x.mean()).fillna(0) 

    # Round the values in the pivot tables to 2 decimal places
    pivot_table_2022 = pivot_table_2022.round(2)
    pivot_table_2023 = pivot_table_2023.round(2)
    fig, axs = plt.subplots(2, 2, figsize=(18, 7), gridspec_kw={'height_ratios': [3, 1]})

    # Plot for 2022 heatmap with reduced height and increased width
    sns.heatmap(pivot_table_2022, cmap='RdYlBu_r', annot=False, fmt=".2f", cbar=True, ax=axs[0, 0],
                 vmin=0, vmax=1, linewidths=3, linecolor='black',linestyle='--', annot_kws={"size": 12}, cbar_kws={'label': 'Contact Rate'})
    axs[0, 0].set_title(f'Contact Rate for Batter {batterID} in 2022', fontsize=16, fontweight='bold', pad=20)
    axs[0, 0].set_xlabel('PlateX ', fontsize=14, labelpad=15)
    axs[0, 0].set_ylabel('PlateZ ', fontsize=14, labelpad=15)
    axs[0, 0].tick_params(axis='both', labelbottom=False, labelleft=False)
    axs[0, 0].grid(True, linestyle='--', linewidth=1, color='black')

    # Plot for 2023 heatmap with reduced height and increased width
    sns.heatmap(pivot_table_2023, cmap='RdYlBu_r', annot=False, fmt=".2f", cbar=True, ax=axs[0, 1],
                 vmin=0, vmax=1, linewidths=3, linecolor='black', linestyle='--',  annot_kws={"size": 12}, cbar_kws={'label': 'Contact Rate'})
    axs[0, 1].set_title(f'Contact Rate for Batter {batterID} in 2023', fontsize=16, fontweight='bold', pad=20)
    axs[0, 1].set_xlabel('PlateX ', fontsize=14, labelpad=15)
    axs[0, 1].set_ylabel('PlateZ ', fontsize=14, labelpad=15)
    axs[0, 1].tick_params(axis='both', labelbottom=False, labelleft=False)
    axs[1, 0].axis('off')
    axs[1, 1].axis('off')
    
    axs[0, 1].grid(True, linestyle='--', linewidth=1, color='black')

    # Show the table for 2022 below the heatmap with rounded values
    table_2022 = axs[1, 0].table(cellText=pivot_table_2022.values, rowLabels=pivot_table_2022.index, 
                                 colLabels=pivot_table_2022.columns, cellLoc='center', loc='center')
    table_2022.auto_set_font_size(False)
    table_2022.set_fontsize(12)
    table_2022.scale(1.5, 1.5)

    # Show the table for 2023 below the heatmap with rounded values
    table_2023 = axs[1, 1].table(cellText=pivot_table_2023.values, rowLabels=pivot_table_2023.index, 
                                 colLabels=pivot_table_2023.columns, cellLoc='center', loc='center')
    table_2023.auto_set_font_size(False)
    table_2023.set_fontsize(12)
    table_2023.scale(1.5, 1.5)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()

batter_ids = data_new['BatterID'].unique()
batter_dropdown = widgets.Dropdown(
    options=batter_ids,
    value=1,
    description='BatterID:',
    disabled=False
)

widgets.interact(plot_heatmaps, batterID=batter_dropdown)
```
